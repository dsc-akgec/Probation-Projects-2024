**Outliers Removal**

```python
Q1 = df.win_by_runs.quantile(0.25)
Q3 = df.win_by_runs.quantile(0.75)
Q1, Q3
```

```python
IQR = Q3-Q1
IQR
```

```python
lower_limit = Q1 - 1.5*IQR
upper_limit = Q3 + 1.5*IQR
lower_limit, upper_limit
```

```python
no_outliers=df[(df.win_by_runs>lower_limit)&(df.win_by_runs<upper_limit)]
```

```python
no_outliers.describe()
```

**Bar Charts**

```python
plt.figure(figsize=(10,6))
sns.countplot(x='winner', data=df)
plt.xlabel('Teams')
plt.ylabel('No. of matches won')
plt.title('Bar Chart of No. of Matches Won vs Teams')
plt.xticks(rotation=90)
plt.show()
```

```python
plt.figure(figsize=(10,6))
sns.countplot(x='city', data=df)
plt.xlabel('Cities')
plt.ylabel('No. of matches played')
plt.title('Bar Chart of No. of Matches Played vs Cities')
plt.xticks(rotation=90)
plt.show()
```

```python
plt.figure(figsize=(7,5))
sns.countplot(x='toss_decision', data=df)
plt.xlabel('Toss Decision')
plt.ylabel('No. of Winners')
plt.title('Bar Chart of No. of Winners vs Toss Decision')
plt.show()
```

**Pie Charts**

```python
a=df.groupby('toss_decision').size()
```

```python
fw = df[df['toss_decision'] == 'field']
twf = fw['winner'].value_counts()
plt.figure(figsize=(7, 7))
plt.pie(twf, labels=twf.index, autopct='%1.1f%%',startangle=-13)
plt.title('Teams Winning After Choosing to Field')
plt.axis('equal')
plt.show()
```

```python
bw = df[df['toss_decision'] == 'bat']
twb = bw['winner'].value_counts()
plt.figure(figsize=(7, 7))
plt.pie(twb, labels=twb.index, autopct='%1.1f%%',startangle=30)
plt.title('Teams winning after choosing to bat')
plt.axis('equal')
plt.show()
```

```python
a=df.groupby('dl_applied').size()
```

```python
plt.figure(figsize=(5,5))
plt.pie(a.values, labels=('NO','YES'), autopct='%1.1f%%',radius=0.8, startangle=90)
plt.title('Pie chart for DLS Applied')
plt.show()
```

```python
b=df.groupby('result').size()
```

```python
plt.figure(figsize=(5,5))
plt.pie(b.values, labels=('No Result','Normal Result','Tie'), autopct='%1.1f%%',radius=1, startangle=0,explode=(0.3,0.5,0.1))
plt.title('Pie chart for Different Types of Results')
plt.show()
```

```python
data = df.groupby(['city', 'winner']).size().unstack(fill_value=0)
plt.figure(figsize=(12, 8))
sns.heatmap(data, cmap="coolwarm", linewidths=0.5,annot=True)
plt.title('Heatmap of City vs. Winning Team')
plt.xlabel('Winning Teams')
plt.ylabel('Cities')
plt.show()
```

```python
data1 = df.groupby(['toss_winner', 'winner']).size().unstack(fill_value=0)
plt.figure(figsize=(10, 8))
sns.heatmap(data1, cmap="coolwarm", linewidths=.5,annot=True)
plt.title('Heatmap of Toss Winner vs. Match Winner')
plt.xlabel('Match Winner')
plt.ylabel('Toss Winner')
plt.tight_layout()
plt.show()
```

**Histogram**

```python
plt.figure(figsize=(10, 6))
plt.hist(df['win_by_runs'], bins=20, edgecolor='black')
plt.title('Distribution of Wins by Runs')
plt.xlabel('Win by Runs')
plt.ylabel('Frequency')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
plt.hist(df['win_by_wickets'], bins=20, edgecolor='black')
plt.title('Distribution of Wins by Wickets')
plt.xlabel('Win by Wickets')
plt.ylabel('Frequency')
plt.show()
```
